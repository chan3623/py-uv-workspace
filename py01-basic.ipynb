# 자료형


1. 데이터 저장방법에 따른 자료형의 분류  
  직접(direct)형 : 직접 데이터를 표현하는 자료형 : int, float, complex      
  시퀀스(sequence)형: 다른 데이터를 포함하는 자료형, 순서가 있는 집합 자료형 : list,str, tuple, bytes, bytearray,range    
  매핑(Mapping)형 : 다른 데이터를 포함하는 자료형 : dict    
  집합(Set)형 : 순서가 없고, 중복된 값도 없다. : set, frozenset  


2. 변경 가능성에 따른 자료형 분류  
  변경 가능형(Mutable) : 데이터의 값을 변경할 수 있다.  : list, dict, set  
  변경 불가능형(Immutable) : 데이터의 값을 변경할 수 없다. : int, float, complex, str, tuple, frozenset


3. 데이터 저장 개수에 따른 자료형의 분류  
  리터널(Literal/Scalar)형 : 한 가지 객체만 저장한다. : int, bytes, str, float, complex  
  컨테이너(Container)형 : 여러 가지 객체를 저장한다. : list, tuple, dict, set, fronzenset

```python
# indexing(인덱싱)

s_val = "hello python"

print(s_val[1])
```

```python
# slicing(슬라이싱) [start : end : step] :step은 생략 가능, step default 1

print(s_val[1:3]) # 1부터 3 이전까지

print(s_val[0:]) # 0부터 끝까지

print(s_val[:len(s_val)]) # 0부터 끝까지

print(s_val[-1]) # 마지막 index

print(s_val[-3 : -1]) # 뒤에서 3번째 index부터 마지막 이전까지

print(s_val[-1 : -3]) # step이 생략으로 인해 1이고 마지막 index에서 시작했기 때문에 아무것도 나오지 않는다.

print(s_val[-1 : -3 : -1]) # step을 -1로 하여 뒤에서 앞으로 가기 때문에 마지막 index에서 뒤에서 3번째 이전까지
```

```python
# str과 int는 연산 불가능

#s_val + 2 -> # TypeError: can only concatenate str (not "int") to str

# 단 *는 가능 (문자열을 반복하라)

s_val * 2

# 문자열 외 다른 타입과만 * 연산자로 반복 가능(문자열끼리는 안됨)

# s_val * s_val # TypeError: can't multiply sequence by non-int of type 'str'
```

문자열 포매팅(서식지정) : 어떤 문서 양식을 만들어 넣고 빈칸에 필요한 내용을 채워서 문서를 지정한다.


문자열포매팅(서식지정) 형식


|코드|설명 |
|---|:---:|
|%d|정수|
|%f|실수|
|%s|문자열|
|%c|하나의문자|
|%o|8진수|
|%x|16진수|
|%%|%기호자체를출력하고자할때|


```python
s_val7 = 'python'
s_val8 = 83.5


print('%s 의 점수는 %.1f입니다.' % ( s_val7, s_val8)) # python1, 2방식
print('{} 의 점수는 {}입니다.'.format(s_val7, s_val8)) # python3 방식
print('{1} 의 점수는 {0}입니다.'.format(s_val8, s_val7))
print('{subject} 의 점수는 {jumsu}입니다.'.format(jumsu=s_val8, subject=s_val7))
print('{{0}} 의 점수는 {{1}}입니다.'.format(s_val7, s_val8)) # { }자체를 출력할때 {{}}을 한다.
print('{{{0}}} 의 점수는 {{{1}}}입니다.'.format(s_val7, s_val8)) # {}자체와 값을 출력할때는 {{{}}}을 한다.


print('{0:<10} 의 점수는 {1}입니다.'.format(s_val7, s_val8))  # 10자리 확보하고 남은 공간은 오른쪽을 여백으로 채운다.
#print('{:<10} 의 점수는 {}입니다.'.format(s_val7, s_val8))


print('{0:>10} 의 점수는 {1}입니다.'.format(s_val7, s_val8))
```

List는 한 변수에서 데이터를 각각 꺼내는 언팩킹(unpacking) 지원을 한다.

```python
# *은 어디에서 사용해도 상관 없지만 한번만 사용 가능

fruits = ["apple", "mange", "grape", "banana"]

print(fruits)

zero, one, _, _ = fruits

print(zero, one, _)

zero, *group = fruits

print(zero, group)

*group, zero = fruits

print(group, zero)

zero, *grape, three = fruits

print(zero, *grape, three)

print(type(zero), type(group))
```

## dict
- 특정 키를 주면 이와 관련된 값을 돌려주는 내용 기반으로 검색하는 자료형이다.
- 데이터의 저장 순서가 없다.
- 내부적으로 해시 기법을 사용하기 때문에 검색 속도가 빠르다.
- key-value형식으로 저장을 하기 때문에 key을 이용해서 value에 접근한다.
- key 값은 중복을 허용하지 않는다.

```python
d = {}

print(d, type(d), dir(d))
```

```python
d = {"one" : 1, "two" : 2}

print(d, d["one"])
```

```python
d = dict(one=1, two=2)

print(d, type(d), d["one"])
```

zip() 함수를 이용해서 dict() 생성할 수 있다

```python
keys = ["one", "two", "three"]
values = [1, 2, 3]

a = zip(keys, values)

print(a)

d = dict(a)

print(d, d["one"])
```

```python
d = {"one" : 1, "two" : 2}

print(d.values())

for k in d.values():
    print(k)
```

```python
d = {"one" : 1, "two" : 2}

print(d.items())

for k, v in d.items():
    print(k, v)
```
