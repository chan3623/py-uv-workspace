[문제1]  
문자열 하나를 입력받아 인터넷 주소를 반환하는 make_url 함수를 정의하세요.    
make_url("naver")  

[출력결과]    
www.naver.com  


```python
def make_url(address):
    if address:
        return "www." + address + ".com"

print(make_url("naver"))
```

[문제2]   
문자열을 입력받아 각 문자들로 구성된 리스트로 반환하는 make_list 함수를 정의하세요.  
make_list("abcd")  

[출력결과]

['a', 'b', 'c', 'd']


```python
def make_list(str):
    if str:
        return list("".join(str))

print(make_list("abcd"))
```

[문제3]   
숫자로 구성된 하나의 리스트를 입력받아, 짝수들을 추출하여 리스트로 반환하는 pickup_even 함수를 구현하세요.   

pickup_even([3, 4, 5, 6, 7, 8])

[출력결과]   
[4, 6, 8]

```python
def pickup_even(nums):
    return list(filter(lambda x : x % 2 == 0, nums))

print(pickup_even([3,4,5,6,7,8]))
        
```

[문제4]  
콤마가 포함된 문자열 숫자를 입력받아 정수로 변환하는 convert_int 함수를 정의하세요.

convert_int("1,234,567")

[출력결과]   
1234567

```python

def convert_int(str):
    return int(str.replace(",", ""))

print(convert_int("1,234,567"))
```

[문제4]  
세 개의 숫자를 입력받아 가장 큰수를 출력하는 print_max 함수를 정의하라. 단 if 문을 사용해서 수를 비교하세요.   
max(10, 2, 5)

[출력결과]  
10

```python
def print_max(*nums):
    num1, num2, num3 = nums

    if num1 >= num2 and  num1 >= num3:
        return num1
    elif num2 >= num1 and  num2 >= num3:
        return num2
    else:
        return num3
    
print(print_max(10, 2, 5))
```

[문제5]    
2 와 5의 배수가 아닌 갯수와 합계를 구하는 함수를 구현하세요.    

resultr(1, 3, 5, 8, 9, 11, 15, 19, 18, 20, 30, 33, 31)  

[출력결과]    
2 와 5의 배수가 아닌 갯수 : 7    
2 와 5의 배수가 아닌 합계 : 107   

==============================================  
result(5,2,7,9,10,8,6)     

[출력결과]    
2 와 5의 배수가 아닌 갯수 :  2
2 와 5의 배수가 아닌 합계 :  16



```python
def resultr(*nums):
    result = []
    for num in nums:
        if num % 2 != 0 and num % 5 != 0:
            result.append(num)
    print(result)
    print(f"2와 5의 배수가 아닌 갯수 : {len(result)}")
    print(f"2와 5의 배수가 아닌 합계 : {sum(result)}")

resultr(1, 3, 5, 8, 9, 11, 15, 19, 18, 20, 30, 33, 31)
resultr(5,2,7,9,10,8,6)
```

```python
# 문제6)
def calc(a, b, /, c, *, d):
    return a + b + c + d
```

```python
# 문제7)
def introduce(name, /, age, *, country):
    return f"{name}, {age} years old from {country}"
```

```python
# 문제8)
def f(x, *args, y, **kwargs):
    return x, args, y, kwargs
```

```python
print(f("x", "a", "r", "g", "s", y="y", name="홍길동", age=20))
```

```python
# 문제9)
def display(*args, **kwargs):
    print("ARGS:", args)
    print("KWARGS:", kwargs)
```

```python
display("a", "r", "g", "s", name="홍길동", age=20)
```
