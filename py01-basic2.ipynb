```python
a, *b = 1, 2, 3, 4
```

```python
a
```

```python
b
```

```python
# *b = 1, 2, 3, 4 # SyntaxError: starred assignment target must be in a list or tuple
```

콤마(,)를  붙여주면 전체를 나머지로 받아올 수 있다.

```python
b
```

### 조건/판별

논리연산자 (and/or)

```python
a = 1
```

```python
a > 2
```

```python
a < 4
```

and 앞에서 True이면 뒤에 것을 반환하고,  
앞에서 False이면 앞에 것을 반환한다.

```python
a > 2 and a < 4
```

or는 앞에서 True이면 앞에 것을 반환하고,  
앞에서 False이면 뒤에 것을 반환한다.

```python
a > 2 or a < 4
```

영문자일 때 공백 포함되어 있어도 interning 기법을 사용할 수 있도록 intern을 사용하면 된다.

```python
from sys import intern
```

```python
c = intern("python programming")
d = intern("python programming")
```

```python
c == d
```

```python
c is d
```

```python
# 숫자일 때는 intern()을 사용할 수 없다.

# a = intern(255) #TypeError: intern() argument must be str, not int
# b = intern(255)
```
