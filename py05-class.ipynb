### staticmethod
- 첫 번째 인자로 self, cls를 받지 않는다.
- 클래스나 인스턴스와 관계없는 일반적인 유틸리티 함수를 정의할때 사용한다.
- 독립적인 기능이지만 클래스 안에 넣는 것이 구조적으로 더 나을때 사용한다.


| 항목         | classmethod             | staticmethod |
| ---------- | ----------------------- | ------------ |
| 첫 번째 인자    | `cls` (클래스)             | 없음           |
| 클래스 변수 접근  | 가능                      | 불가능          |
| 인스턴스 변수 접근 | 불가능                     | 불가능          |
| 주 용도       | 클래스 상태를 사용하는 메서드        | 유틸리티 함수 정의   |
| 호출 방식      | 클래스.메서드() 또는 인스턴스.메서드() | 동일           |


````python
class B:
  x = 1


  def xx(self, x): # 인스턴스 메소드
    self.x = x


  @classmethod
  def yy(cls, x):  # 클래스 메소드
    cls.x = x


  @staticmethod
  def zz(a, b):  # 스테이틱 메소드
    print(a+b)


  @staticmethod  # 스테이틱 메소드
  def kk():
    print(x)
````

````python
B.zz(3, 4)
````

````python
B.kk()
````

[실행결과]  
피자의 면적은 314.0  
도넛의 면적은 12.56 


````python
class Circle:
    def setArea(self, name, radius):
        self.name = name
        self.radius = radius

    def getArea(self):
        return 3.14 * self.radius * self.radius

ca = Circle()
ca.setArea('피자', 10)

cb = Circle()
cb.setArea('도너츠', 2)

print("{}의 면적은 {}입니다. ".format(ca.name, ca.getArea()))
print("{}의 면적은 {}입니다. ".format(cb.name, cb.getArea()))

class Circle:
    @staticmethod
    def setArea(name, r):
        return f"{name}의 면적은 {r * r * 3.14}입니다."

print(Circle.setArea("피자", 10))
print(Circle.setArea("도넛", 2))
````

````python
class Dog:
    count = 0

    def __init__(self, name):
        self.name = name
        Dog.count += 1

    @classmethod
    def get_count(cls):
        return cls.count

    @staticmethod
    def is_valid_name(name):
        return isinstance(name, str) and len(name) > 0
````

````python
from abc import ABC, abstractmethod

class Shape(ABC):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    @abstractmethod
    def area(self):
      pass    

class Rect(Shape):
    def __init__(self, width, height ):
        super().__init__(width, height)

    def area(self):
      return self.width * self.height  
   
class Triangle(Shape):
    def __init__(self, width, height ):
        super().__init__(width, height)

    def area(self):
      return self.width * self.height * 0.5
````

````python
r = Rect(2, 5)
print(f"사각형 넓이 : {r.area()}")
````

````python
t = Triangle(2, 5)
print(f"삼각형 넓이 : {t.area()}")
````
