## 예외처리
- 예외처리는 에러가 발생하되 실행을 중단하지 않고 계속 실행하게 해주는 방법이다.
- 예외처리 구조

```
try:
    실행할 코드
except:
    오류가 발생했을 때 처리하는 코드
else:
    오류가 없을 때 처리하는 코드
finally:
    무조건 처리하는 코드
```

- 파이썬에서는 선언할 때 오류를 일일이 체크하지 안혹, 사용할 ㄹ때 에러가 발생하면 디버깅하는 스타일이다.
- 허락보다는 용서를 구하기가 쉽다.(Easier to ask for forgiveness than permission)
- except 뒤에 에러 이름을 넣으면, 해당하는 에러가 발생했을 때만 except를 발생시킨다.
- Built-in Exceptions 예외 계층도 : 
    - https://docs.python.org/ko/3/tutorial/errors.html
    - https://docs.python.org/3/library/exceptions.html


````python
# try 블록이 정상적으로 실행되지 않으면 else 블록은 실행되지 않는다.
# try 블록에서 예외가 발생했을 때 : try 실행 => except 실행 => finally 실행
# try 블록에서 정상적으로 실행됐을 때 : try 실행 => else 실행 => finally 실행

try:
    a = int('3')
    b = int('5')
    c = int('4')
except:
    print('숫자가 아닌 문자가 입력되었습니다.')
else:
    print(f"{a}와 {b}와 {c}의 합 : {a + b + c}")
finally:
    print("program end")
````

## 다중 except
- 다중 except 선언이 가능하다.
- except뒤에 오는 as는 try 블록에서 발생한 이름을 받아온다.
```
  try:
     실행할 코드
  except 예외 as 변수:
     예외가 발생했을 때 처리하는 코드
  except 예외 as 변수:
     예외가 발생했을 때 처리하는 코드  
  except:
     예외가 발생했을 때 처리하는 코드  
```

````python
y = [10, 20, 30]

try:
    index, x = map(int, input("인덱스와 나눌 숫자를 입력하세요.").split())
    print(y[index] // x)
except ZeroDivisionError as e:
    print("숫자를 0으로 나눌 수 없습니다.", e)
except IndexError as e:
    print("잘못된 인덱스입니다.", e)
except Exception as e:
    print(e)
else:
    print("정상처리 되었습니다.")
finally:
    print("program end")
````

````python
class Bird:
    def fly(self):
        raise NotImplementedError # 상속받은 자식클래스에서 fly()을 오버라이딩을 하지 않고 사용할 시 에러 발생

class Eagle(Bird):
    def display(self):
        print("display")

    def fly(self): # fly()를 오버라이딩 했기 때문에 에러가 발생하지 않는다.
        print("very fast")

eagle = Eagle()
eagle.display()
eagle.fly()
````

````python
# 추상 클래스일 경우 @abstractmethod로 지정한 메소드를 자식 클래스에서 오버라이딩하지 않을 경우 에러를 발생 

from abc import ABC, abstractmethod

class Bird(ABC):
  @abstractmethod
  def fly(self):
    pass


class Eagle(Bird):
  def display(self):
    print('display')


# TypeError: Can't instantiate abstract class Eagle without an implementation for abstract method 'fly'

  def fly(self): 
    print('very fast')


eagle = Eagle()
eagle.fly()
````

## 사용자 정의 예외 클래스

````python
class UserClass(Exception):
    def __str__(self):
        return "허용하지 않는 데이터입니다."
````

````python
try:
    x = int(input("숫자 입력 : "))
    if x <= 0:
        raise UserClass
    print(x)
except UserClass as e:
    print(e)
except ValueError as e:
    print(e)
except Exception as e:
    print(e)
except BaseException as e:
    print(e)
````

````python
# return이 있어도 finally는 무조건 실행

def test():
    try:
        print("try 실행1")
        return 
        print("try 실행2")
    except Exception as e:
        print('except', e)
    else:
        print("else")
    finally:
        print("finally")

test()
````
